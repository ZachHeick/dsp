## Q16:   Create a horizontal frequency barplot for 'region'

````python
s = df.region.value_counts(sort=True, ascending=True)
values = s.values.tolist()
regions = s.keys().tolist()

y_pos = np.arange(len(regions))

plt.figure(figsize = (10,5))
plt.barh(y_pos, values, align='center')
plt.yticks(y_pos, regions)
plt.ylabel('Region')
plt.xticks(np.arange(0,400,50))
plt.xlabel('Frequency')
plt.title('Frequency of Regions')
plt.grid(axis='x')
plt.show()
````

## Q17:   Create a horizontal frequency barplot for 'sub_region'

````python
s = df.sub_region.value_counts(sort=True, ascending=True)
values = s.values.tolist()
regions = s.keys().tolist()

y_pos = np.arange(len(regions))

plt.figure(figsize = (10,5))
plt.barh(y_pos, values, align='center')
plt.yticks(y_pos, regions)
plt.ylabel('Sub-Region')
plt.xticks(np.arange(0,250,25))
plt.xlabel('Frequency')
plt.title('Frequency of Sub-Regions')
plt.grid(axis='x')
plt.show()
````
